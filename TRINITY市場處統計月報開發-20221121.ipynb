# 臺、韓、日、中及新加坡對主要市場市場出口成長率比較

```python
df2 = loading_main_competer_data()
df2.tail(20)
```

```python
# setting target month  設定2022年1到table5_setting_target_month的月份區間
if setting_target_month > 1:
    table5_setting_target_month = setting_target_month - 1
else:
    table5_setting_target_month = 1
#table5_setting_target_month = 
print("table5_setting_target_month: {}".format(table5_setting_target_month))

##output each major competitor export condition
part_taiwan = table_taiwan(df2, TABLE5_COUNTRY_ITEM_MAP[0][0],TABLE5_COUNTRY_ITEM_MAP[0][1], setting_target_year, table5_setting_target_month, TABLE5_COL_ITEM)
part_korea = table_korea(df2, TABLE5_COUNTRY_ITEM_MAP[1][0],TABLE5_COUNTRY_ITEM_MAP[1][1], setting_target_year, table5_setting_target_month, TABLE5_COL_ITEM)
part_japan = table_japan(df2, TABLE5_COUNTRY_ITEM_MAP[2][0],TABLE5_COUNTRY_ITEM_MAP[2][1], setting_target_year, table5_setting_target_month, TABLE5_COL_ITEM)
part_china = table_china(df2, TABLE5_COUNTRY_ITEM_MAP[3][0],TABLE5_COUNTRY_ITEM_MAP[3][1], setting_target_year, table5_setting_target_month, TABLE5_COL_ITEM)
#part_china
table5 = part_taiwan.merge(part_korea,on='市場', how='left')
table5 = table5.merge(part_japan, on='市場', how='left')
table5 = table5.merge(part_china, on='市場', how='left')
table5
```
